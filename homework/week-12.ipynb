## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt
slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]

plt.figure(figsize=(10, 10))
plt.pie(slices, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart: Country Distribution')

plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)

plt.figure(figsize=(10, 6))
plt.hist(n, bins=30, edgecolor='black')

plt.title('Histogram: Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data Points')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label='Trend Line')

plt.title('Scatter Plot (With Trend Line)')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.legend()

df = pd.DataFrame({'x': x, 'y': y})

plt.show()
```
